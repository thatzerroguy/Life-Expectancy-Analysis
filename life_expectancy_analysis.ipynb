## Finding any Missing value

Here we can see that there are missing values in the country and hale_birth_both columns

```python
life_data.isna().sum()
```

```python
life_data[life_data['country'].isna()]
```

A trend i noticed was that every country appeared in four years which this country did not so i filled it with the missing value

```python
life_data['country'].fillna("Lao People's Democratic Republic", inplace=True)
```

```python
life_data.isna().sum()
```

```python
life_data[life_data['hale_birth_both'].isna()]
```

I noticed that the ages for hale_birth_both was the average of the two genders so i found the average and filled in the missing values

```python
average_hale_birth_male_female = (life_data['hale_birth_male'] + life_data['hale_birth_female']) / 2
```

```python
life_data['hale_birth_both'].fillna(average_hale_birth_male_female, inplace=True)
```
